## 1. Generate a fully-textured 3D Mesh

To generate a 3D model you have to use **create_3dmodel()** method. It can be used with a lot of optional argument listed below.

> **Optional Arguments:**
    > - **outdirname:** The name of the project
> - **inpath:** You can pass the input image path directly. If you don't use it, it will be asked by the method.
    > - **scale:** Image scaling, from 0% to 100%.
    > - **outpath:** The folder where the project is going to be stored.
    > - **s3path:** Only needed if s3 is not in the default path.
    > - **gray:** Grayscale, True or False.
    > - **sample:** Mesh reconstruction from a Artemis sample, True or False. 
    > - **reslevel:**    

```python
proj_path = phg.create_3dmodel(inpath = "",
                               scale = 1, gray = False, sample = True, reslevel = 3)
```
